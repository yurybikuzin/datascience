Избавьтесь от выбросов, чтобы они не ухудшили качество нашей модели. Для того, чтобы избавиться от выбросов, уберите все данные, значения признаков которых отстают от первого или третьего квартиля более, чем на 1.5 межквартильного размаха.

```python
import pandas as pd
datah = pd.read_csv("./data/heart_fin.xls", sep=';', low_memory = False)
datah.target.value_counts()
```

```python
def in_quantile(df):
    quartile_1 = df.quantile(0.25)
    quartile_3 = df.quantile(0.75)
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - iqr * 1.5
    upper_bound = quartile_3 + iqr * 1.5
    return df[~((df > upper_bound) | (df < lower_bound)).any(axis=1)]
# data = datah.drop(['fbs'], axis=1)
in_quantile(datah).shape
datah = in_quantile(datah)
```

```python
X = datah[list(set(datah.columns) - set(['target']))]
y = datah.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="lbfgs", max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
fpr, tpr, thresholds = roc_curve(y_test, y_pred, drop_intermediate=False)
(fpr, tpr, thresholds)
roc_auc = roc_auc_score(y_test, y_pred)

plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.title('Receiver Operating Characteristic', fontsize=15)
plt.xlabel('False positive rate (FPR)', fontsize=15)
plt.ylabel('True positive rate (TPR)', fontsize=15)
plt.legend(fontsize=15)
roc_auc
```

# Решение ментора:
    

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
heart_data = pd.read_csv('./data/heart_fin.xls', sep = ";")
Q1 = heart_data.quantile(0.25)
Q3 = heart_data.quantile(0.75)
IQR = Q3-Q1
heart_data = heart_data[~((heart_data < (Q1 - 1.5 * IQR)) |(heart_data > (Q3 + 1.5 * IQR))).any(axis=1)]
print(heart_data.shape)
X = heart_data.drop(columns=['target'])
y = heart_data['target']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.15, random_state=5)
model = LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)
Y_predicted = model.predict(X_valid)
print("ROC AUC score {}".format(roc_auc_score(y_valid,Y_predicted)))
```
